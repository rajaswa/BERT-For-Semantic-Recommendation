```python
import pandas as pd
import numpy as np

from nltk.cluster import KMeansClusterer
import nltk
  
from sklearn import cluster
from sklearn import metrics

from utils import *
```

```python
query = "A book about murder mystery involving a thrilling romantic adventure in the city of Paris"
```

```python
df = pd.read_pickle('cmu_data_phrase.pkl')
df.head()
```

```python
#Get top-n entries by score ranking
df['score'] = df.apply(lambda row: get_score(query, row.noun_phrase, row.sentiment, alpha=0.95), axis = 1) 
df_topn = df.nlargest(500, 'score')
df_topn.reset_index(inplace=True)
```

```python
recommendations = get_recommendation(query, df_topn, n=500, NUM_CLUSTERS=10)
```

```python
for book in recommendations:
    print("Book Title : ", book['book_title'])
    print()
    print("Book Author : ", book['book_author'])
    print()
    print("Book Summary :")
    print(book['summary'])
    print()
    print("####################################")
    print()
```
